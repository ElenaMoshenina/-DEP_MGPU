```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Загружаем данные из CSV-файла
survey_data = pd.read_csv('data.csv')
```

```python

# Исключаем первый столбец (время прохождения опроса)
survey_data = survey_data.iloc[:, 1:]
```

```python
#Функции для анализа вопросов с разными типами ответов

# Функция для анализа вопросов с одним вариантом ответа
def evaluate_single_answer(responses, question_title):
    # Подсчитываем частоту каждого ответа
    response_counts = responses.value_counts()
    print(f"\nАнализ вопроса: {question_title}")
    print(response_counts)

    # Визуализация данных
    plt.figure(figsize=(8, 6))
    plt.barh(response_counts.index, response_counts.values, color=plt.cm.Paired.colors)
    plt.title(question_title)
    plt.xlabel('Количество ответов')
    plt.ylabel('Варианты ответов')
    plt.show()

# Функция для анализа вопросов с множественным выбором
def evaluate_multiple_answers(responses, question_title):
    # Разделяем ответы по точке с запятой и создаем список всех ответов
    all_responses = responses.dropna().str.split(';').explode()
    
    # Подсчитываем частоту каждого ответа
    response_counts = all_responses.value_counts()
    print(f"\nАнализ вопроса: {question_title}")
    print(response_counts)

    # Визуализация данных
    plt.figure(figsize=(8, 6))
    plt.barh(response_counts.index, response_counts.values, color=plt.cm.Set2.colors)
    plt.title(question_title)
    plt.xlabel('Количество ответов')
    plt.ylabel('Варианты ответов')
    plt.show()

# Функция для анализа вопросов с шкалой
def evaluate_scale(responses, question_title):
    mean_value = responses.mean()
    distribution = responses.value_counts().sort_index()
    print(f"\nАнализ вопроса: {question_title}")
    print(f"Среднее значение: {mean_value:.2f}")
    print(distribution)

    plt.figure(figsize=(8, 6))
    plt.hist(responses, bins=len(distribution), edgecolor='black', color='skyblue')
    plt.title(question_title)
    plt.xlabel('Оценка')
    plt.ylabel('Количество ответов')
    plt.xticks(sorted(responses.unique()))
    plt.show()
```

```python
# Функция для автоматического анализа всех, чтобы не пришлось прописывать каждый вопрос вручную
def process_survey(data):
    for column in data.columns:
        print(f"\n=== Анализ вопроса: {column} ===")
        
        # Определяем тип вопроса
        if data[column].dtype == 'object':  # Если данные текстовые
            # Проверяем, есть ли разделители (точка с запятой) для множественного выбора
            if data[column].dropna().str.contains(';').any():
                evaluate_multiple_answers(data[column], column)
            else:
                evaluate_single_answer(data[column], column)
        elif pd.api.types.is_numeric_dtype(data[column]):  # Если данные числовые
            evaluate_scale(data[column], column)
        else:
            print("Неизвестный тип данных для анализа.")

# Запускаем автоматический анализ
process_survey(survey_data)
```
